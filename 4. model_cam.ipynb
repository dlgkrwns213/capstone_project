```python
import numpy as np
import cv2
import mediapipe as mp
import sys
import os
import pandas as pd

from tensorflow.keras.models import load_model

from multimodel import MultiModelDetector

def split_X(X: np.ndarray):
    return X[:, :15], X[:, 15:30], X[:, 30:46], X[:, 46:62], X[:, 62:74], X[:, 74:76]
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# 버튼 목록
button_names = ['rollhand', 'rollfeet', 'hooray', 'jumphooray', 
                'clap', 'cheek', 'RPS', 'thumbsup', 'all']

# 변수 a 초기값 설정
behavior = 'None'

# 버튼 생성
buttons = [widgets.Button(description=name) for name in button_names]

# 버튼 클릭 이벤트 핸들러 함수
def on_button_click(b):
    global behavior
    behavior = b.description
    with output:
        clear_output(wait=True)
        print(f"{b.description} 버튼이 클릭되었습니다!")
        print(f"현재 behavior: {behavior}")

# 버튼과 이벤트 핸들러 함수 연결
for button in buttons:
    button.on_click(on_button_click)

output = widgets.Output()

# 버튼을 2 * 4 그리드 형태로 배치
buttons_grid = widgets.GridBox(buttons, layout=widgets.Layout(grid_template_columns="repeat(4, 150px)"))

# 버튼과 결과를 노트북에 표시
display(buttons_grid)
display(output)
```

```python
model_version = '_small'

mmd = MultiModelDetector(behavior)

model = load_model(f'models/{behavior}_model{model_version}.keras')

# 카메라 캡처를 위한 객체 생성
cap = cv2.VideoCapture(0) 

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Convert frame to angles
    datas = mmd.image_to_numpy(frame)

    # Make prediction using the model
    prediction = model.predict(split_X(datas.reshape(1, -1)))

    # # Get predicted beh
    predicted_behavior_index = np.argmax(prediction)
    predicted_behavior = mmd.behaviors[predicted_behavior_index]

    # # Display the frame and predicted behavior
    cv2.putText(frame, predicted_behavior, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    showed = mmd.find_body(frame)
    cv2.imshow('frame', showed)

    # Quit if 'q
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```
