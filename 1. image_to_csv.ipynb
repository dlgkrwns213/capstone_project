```python
import numpy as np
import pandas as pd
import os
import cv2
import csv

import multimodel
from multimodel import MultiModelDetector
```

```python
mmd = MultiModelDetector('find_image')
zeros_array = np.zeros((91, 3))

total_images_directory = 'datasets_images'
total_csvfile_directory = 'datasets_locations_csvs'
if not os.path.exists(total_csvfile_directory):
    os.makedirs(total_csvfile_directory)
    
for behavior_image_directory in os.listdir(total_images_directory):
    print(f'{behavior_image_directory} directory start!')
    behavior_image_directory_path = os.path.join(total_images_directory, behavior_image_directory)

    image_datas = []
    for image_directory in os.listdir(behavior_image_directory_path):
        image_directory_path = os.path.join(behavior_image_directory_path, image_directory)
        for filename in os.listdir(image_directory_path):
            if filename.endswith('.jpg') or filename.endswith('.png'):
                image_path = os.path.join(image_directory_path, filename)
                image = cv2.imread(image_path)
                
                image_data = mmd.image_get_location(image)
                if np.array_equal(image_data, zeros_array) and 'None' not in image_directory:
                    continue
                image_datas.append(image_data)

    print(behavior_image_directory, np.array(image_datas).shape)

    # directory 생성
    csvfile_directory_path = os.path.join(total_csvfile_directory, behavior_image_directory)
    if not os.path.exists(csvfile_directory_path):
        os.makedirs(csvfile_directory_path)

    csvfile_path = os.path.join(csvfile_directory_path, f'{behavior_image_directory}_locations.csv')
    print(f'save csvfile path: {csvfile_path}')
    mode = 'w' # 'a' if os.path.exists(csvfile_path) else 'w'
    with open(csvfile_path, mode, newline='') as csvfile:
        writer = csv.writer(csvfile)
        if mode == 'w':
            # 파일이 존재하지 않으면 헤더를 쓴다
            rh = [f'RightHand{i+1}' for i in range(21)]
            lh = [f'LeftHand{i+1}' for i in range(21)]
            fa = [f'Face{i+1}' for i in range(16)]
            po = [f'Pose{i+1}' for i in range(33)]
            
            writer.writerow(rh + lh + fa + po)
        writer.writerows(image_datas)

    print(f'{behavior_image_directory} directory complete!!\n')
```
